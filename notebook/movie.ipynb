```python
import ast
ast.literal_eval(movies['genres'].iloc[0])

def extract_names(text):
    L = []

    for item in ast.literal_eval(text):
        L.append(item['name'])

    return L
```

### Stemmer

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []

    for word in text.split():
        y.append(ps.stem(word))

    return " ".join(y)
```

```python
version_1['tags'] = version_1['tags'].apply(stem)
```

### CountVectorizer

```python
# Keep only first 5000 movies
movies = movies.iloc[:5000].reset_index(drop=True)

print(movies.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = cv.fit_transform(version_1['tags']).toarray()

print(vectors.shape)
```

```python
vectors.shape
```

## THE Heart

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors) 
```

```python
similarity.shape
```

```python
def recommend(movie):

    movie_index = movies[
        movies['title'].str.lower() == movie.lower()
    ].index[0]

    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:8]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

# FINAL VERSION

```python
# Final model uses Version 3
similarity_final = similarity_3
```

```python
def recommend_final(movie):

    movie_index = movies[
        movies['title'].str.lower() == movie.lower()
    ].index[0]

    distances = similarity_final[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:9]

    recommendations = []

    for i in movies_list:
        recommendations.append(
            movies.iloc[i[0]].title
        )

    return recommendations
```
